### **5. Encode categorical features**



1.   Label Encoding
2.   OneHot Encoding
3.   Target Encoding
4.   Binary Transformation


```python
# One-Hot Encoding
one_hot_encoder = OneHotEncoder(sparse=False)
encoded_cols = one_hot_encoder.fit_transform(merge_df[['StoreType', 'Assortment']])
column_names = one_hot_encoder.get_feature_names_out(['StoreType', 'Assortment'])
encoded_df = pd.DataFrame(encoded_cols, columns=column_names)
print(encoded_df)
```

```python
# Label Encoding
label_encoder = LabelEncoder()

for col in ['StoreType', 'Assortment']:
    merge_df[col] = label_encoder.fit_transform(merge_df[col])

print(merge_df)
```

```python
# Target Encoding
for col in ['StoreType', 'Assortment']:
    target_means = merge_df.groupby(col)['Sales'].mean().to_dict()
    merge_df[col] = merge_df[col].map(target_means)

# Display the DataFrame with target encoding
print(merge_df)
```

```python
# Binary Encoding
for col in ['StoreType', 'Assortment']:
    unique_categories = merge_df[col].unique()
    for category in unique_categories:
        merge_df[f"{col}_{category}"] = (merge_df[col] == category).astype(int)
# Drop the original columns
merge_df.drop(['StoreType', 'Assortment'], axis=1, inplace=True)
print(merge_df)
```

How many days of data are included in the merge_df dataframe, and what is the range of dates?

```python
merge_df['Date'] = pd.to_datetime(merge_df['Date'])

# Calculate the number of unique days
num_unique_days = len(merge_df['Date'].dt.date.unique())

# Find the range of dates
min_date = merge_df['Date'].min().date()
max_date = merge_df['Date'].max().date()

# Print the results
print(f"Number of Days of Data: {num_unique_days}")
print(f"Date Range: {min_date} to {max_date}")
```

What is the average sales per customer for each store in the merge_df dataframe?

```python
# Group the data by 'Store' and calculate the mean of 'Sales' and 'Customers' for each store
average_sales_per_customer = merge_df.groupby('Store')[['Sales', 'Customers']].mean()

# Calculate the average sales per customer by dividing 'Sales' by 'Customers'
average_sales_per_customer['Average Sales per Customer'] = average_sales_per_customer['Sales'] / average_sales_per_customer['Customers']

# Print the average sales per customer for each store
print(average_sales_per_customer[['Average Sales per Customer']])
```

How many days of the week does each store in the merge_df dataframe open and close?

```python
grouped_data = merge_df.groupby('Store')

# Initialize empty dictionaries to store open and close days for each store
open_days_per_store = {}
close_days_per_store = {}

# Iterate through each store group
for store_id, store_data in grouped_data:
    # Extract 'DayOfWeek' and 'Open' columns for the current store
    days_open = store_data[store_data['Open'] == 1]['DayOfWeek'].unique()
    days_closed = store_data[store_data['Open'] == 0]['DayOfWeek'].unique()

    # Store the open and close days in the dictionaries
    open_days_per_store[store_id] = days_open
    close_days_per_store[store_id] = days_closed

# Print the open and close days for each store
for store_id in open_days_per_store:
    print(f"Store {store_id}:")
    print(f"Open on Days: {open_days_per_store[store_id]}")
    print(f"Closed on Days: {close_days_per_store[store_id]}")
    print()
```

What is the correlation between the Sales and Customers columns in the merge_df dataframe?

```python
correlation_sales_customers = merge_df['Sales'].corr(merge_df['Customers'])
print(f"Correlation between Sales and Customers: {correlation_sales_customers:.2f}")
```

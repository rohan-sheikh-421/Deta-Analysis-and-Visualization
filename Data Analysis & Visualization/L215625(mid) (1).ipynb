```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# image
img = cv2.imread('/content/PrimeMinisterNawazSharif.jpg', 0)

#Sobel filter
hx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
hy = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

img_x = cv2.filter2D(img, -1, hx)
img_y = cv2.filter2D(img, -1, hy)


edge_strength = np.sqrt(np.square(img_x) + np.square(img_y))

# Apply threshold
threshold_value = 100
edge_strength[edge_strength < threshold_value] = 0

# Print the images
cv2_imshow( img_x)
cv2_imshow( img_y)
cv2_imshow( edge_strength)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

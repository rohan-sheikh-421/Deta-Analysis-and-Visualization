```python
import pandas as pd
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px

# Load the Netflix dataset
netflix_data = pd.read_csv('/content/netflix_titles.csv')

# Create the Plotly charts
fig1 = px.bar(netflix_data, x='release_year', y='duration', title='Average Duration by Release Year')
fig2 = px.histogram(netflix_data, x='release_year', color='type', barmode='group', title='Count of Movies and TV Shows by Release Year')
fig3 = px.density_heatmap(netflix_data, x='release_year', y='duration', title='Heatmap of Release Year vs. Duration')
fig4 = px.box(netflix_data, x='type', y='duration', points='all', title='Box Plot of Duration by Type')

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the dashboard
app.layout = html.Div(children=[
    html.H1('Netflix Movies and Shows Dashboard', style={'text-align': 'center'}),

    html.Div([
        dcc.Graph(figure=fig1)
    ]),

    html.Div([
        dcc.Graph(figure=fig2)
    ]),

    html.Div([
        dcc.Graph(figure=fig3)
    ]),

    html.Div([
        dcc.Graph(figure=fig4)
    ])
])

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
